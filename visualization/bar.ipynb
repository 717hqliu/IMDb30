```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(16, 6))    # 有两个子图

labels = ['head', 'tail', 'average']
x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

#=================================第一张图：三个模型的 MRR ==================================
MRR_Conv1D = [0.030, 0.072, 0.051]   # Conv1D 模型的 head, tail, 和 average
MRR_Conv2D = [0.014, 0.060, 0.037]
MRR_Conv3D = [0.068, 0.170, 0.119]

bar_MRR_1D = ax0.bar(x - width, MRR_Conv1D, width, label='Conv1D')    # 第一个参数是 bar 的起始位置
bar_MRR_2D = ax0.bar(x,         MRR_Conv2D, width, label='Conv2D')
bar_MRR_3D = ax0.bar(x + width, MRR_Conv3D, width, label='Conv3D')

ax0.set_ylabel('MRR', fontsize=18)
ax0.set_title('MRR of 3 convolutional models', fontsize=18)
ax0.set_xticks(x, labels, fontsize=18)
ax0.legend(fontsize=14)

ax0.bar_label(bar_MRR_1D, padding=3)
ax0.bar_label(bar_MRR_2D, padding=3)
ax0.bar_label(bar_MRR_3D, padding=3)
#==========================================================================================

#=================================第二张图：三个模型的 MR ==================================
MR_Conv1D = [4501, 3779, 4140]   # Conv1D 模型的 head, tail, 和 average
MR_Conv2D = [1634, 2783, 2209]
MR_Conv3D = [1676, 2428, 2052]

bar_MR_1D = ax1.bar(x - width, MR_Conv1D, width, label='Conv1D')    # 第一个参数是 bar 的起始位置
bar_MR_2D = ax1.bar(x,         MR_Conv2D, width, label='Conv2D')
bar_MR_3D = ax1.bar(x + width, MR_Conv3D, width, label='Conv3D')

ax1.set_ylabel('MR', fontsize=18)
ax1.set_title('MR of 3 convolutional models', fontsize=18)
ax1.set_xticks(x, labels, fontsize=18)
ax1.legend(fontsize=13)

ax1.bar_label(bar_MR_1D, padding=3)
ax1.bar_label(bar_MR_2D, padding=3)
ax1.bar_label(bar_MR_3D, padding=3)
#==========================================================================================

fig.tight_layout()
plt.savefig('./MRR-MR.jpg')
plt.show()
```
